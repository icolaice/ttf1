```python
import radiomics
from radiomics import featureextractor
import pandas as pd
```

```python
dataDir = '/Users/icolaice/Desktop/data/CT/'
```

```python
extractor = featureextractor.RadiomicsFeatureExtractor()
df = pd.DataFrame()
```

```python
imageName ='/Users/icolaice/Desktop/data/CT/001/data.nrrd'
maskName ='/Users/icolaice/Desktop/data/CT/001/mask.nrrd'
featureVector = extractor.execute(imageName,maskName)  
df_add = pd.DataFrame.from_dict(featureVector.values()).T
df_add.columns = featureVector.keys()  # 特征的名称
df = pd.concat([df,df_add])
df.to_excel(dataDir + 'results.xlsx')
```
